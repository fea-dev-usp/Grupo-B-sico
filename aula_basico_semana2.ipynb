##### Diferen√ßa de efici√™ncia for e while

```python
def timefor():
    lista = []
    for i in range(100):
        lista.append(i)
    return lista

%timeit timefor()
```

```python
def timewhile():
    lista = []
    x = 0
    while x < 100:
        lista.append(x)
        x += 1
    return lista

%timeit timewhile()
```

```python
np.random.seed(1)
final_tails = []

for x in range(1000) :
    tails = [0]
    for x in range(10) :
        coin = np.random.randint(0, 2)
        tails.append(tails[x] + coin)
    final_tails.append(tails[-1])

plt.hist(final_tails, bins = 10)
plt.show()
```

##### Outros usos do random

```python
s_0 = 100
rets_plus_1 = np.random.normal(loc=(1+0.07)**(1/12), scale=(0.15*np.sqrt(1/12)), size=(12*10, 100))
rets_plus_1[0] = 1
ret_val = s_0*pd.DataFrame(rets_plus_1).cumprod()
print(ret_val)
plt.plot(ret_val)
plt.show()
```
