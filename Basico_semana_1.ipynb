### Variáveis

#### O que são

```python
# Armazenamento de valores

# Atribuindo "2" à variável "x"
x = 2
print(x+2,x*4)
```

```python
# Mesmo processo, "feito à mão"
print(2+2,2*4)
```

```python
# Peguemos um caso mais extremo para analisarmos a praticidade do uso de variáveis
import numpy as np

# Atribuindo o array à variável "arr"
arr = np.array([1.1,1.2,1.3,1.4,1.5])
print(arr+arr,arr*4)
```

```python
# Mesmo processo, "feito à mão"
print(np.array([1.1,1.2,1.3,1.4,1.5])+np.array([1.1,1.2,1.3,1.4,1.5]),np.array([1.1,1.2,1.3,1.4,1.5])*4)
```

### Listas

#### Conceitos básicos

```python
# Serve para agrupar diferentes valores/informações dentro de uma mesma variável

# Processo manual
x1 = 1
x2 = 2
x3 = 3
x4 = 4

# Feito com lista
lista = [1,2,3,4]
```

```python
# Estruturação com colchetes
lista_1 = [2,3,4,5]
lista_2 = list([2,3,4,5])

print(lista_1,lista_2)
```

#### Acesso aos valores 

```python
# Feito a partir do index do valor que queremos acessar
# Contagem feita a partir do 0
print(lista[0])
```

```python
# Index negativo
print(lista[-1])
```

```python
# Index de lista de listas
l1 = [1,2,3,4]
l2 = [2,3,4,5]
lt = list([l1,l2]) # equivalente a lt = [l1,l2]

print(lt)
print(lt[1][2])
```

```python
# Acesso a todos os valores
print(lista[:])

# Acesso a parte dos valores
print(lista[-2:])
```

#### Manipulação 

```python
# Adicionando elementos
l_1 = [5,6,8,9]

# Append (adiciona no final da lista)
l_1.append(6)
print(l_1)

# Insert (permite a escolha do index)
l_1.insert(3,4)
print(l_1)
```

```python
# Removendo elementos
l_2 = [2,4,6,8]

# Pop (remove de um index em particular)
l_2.pop(2)
print(l_2)
```

```python
# Concatenando
print(l_1+l_2)
```

```python
# Ordenação
l_1.sort()
print(l_1)
```

### Funções

#### Estrutura

```python
# Iniciamos com "def", damos o nome para a função e explicitamos nos parênteses os argumentos (nesse caso são obrigatórios)
def soma(a,b):
    s = a + b
    print(s)
    
soma(3,4)
```

```python
# Caso com argumentos opcionais
def soma(a=0,b=1):
    s = a + b
    print(s)
    
soma()
```
